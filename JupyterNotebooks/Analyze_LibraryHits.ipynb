# Summary Tag Information from GNPS Putative Annotations (Library Hits)
Provide context to your GNPS putative annotations from spectral library matching via "GNPS Tags". 

**Steps:**
1. Download the "View All Library Hits" or "View All Spectra Hits" table from your GNPS jobs.
2. Upload the .tsv file to the "input" folder.
2. Update the path to your file.
3. Run the plots_libraryhits function to generate all information *or* run the separate "summary...libraryhits()" functions
4. Download a single file **or** the zip file associated with each type of tag.

The following are compatible: (1) Molecular Networking (classic), (2) Library Search, and (3) Feature-based Molecular Networking.

**More information and documentation** - https://ccms-ucsd.github.io/GNPSDocumentation/tag_info/

*you can run this notebook with the provided example file*
