# Summary Tag Information After Completion of the Tag Template - BETA
Using the tag template for analysis will provide the most complete (as you have likely entered tags and/or InChIKeys) and up-to-date information for your own data. 

*The information contained in your template will eventually be merged into the Mastersheets by admins; however, this process will take time and releases will be cut periodically.*

**Steps:**
1. Upload your tag template to the "input" folder.
2. Update the path to your file
3. Run the plots_tagtemplate function to generate all information *or* run the separate "summary...tagtemplate()" functions
4. Download a single file **or** the zip file associated with each type of tag.

The following are compatible: (1) Molecular Networking (classic), (2) Library Search, and (3) Feature-based Molecular Networking.

**More information and documentation** - https://ccms-ucsd.github.io/GNPSDocumentation/tag_info/
